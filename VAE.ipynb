```python
from ProcessData.TrainingLoss import TrainingLoss
from VAE.DataBase import DataBase
from VAE.Train import TrainingConfig, Train
from VAE.LossType import VAELossType
import Format
import time

name = Format.name
DB_train = DataBase('TrainingData')
DB_validate = DataBase('ValidationData')

error_config = TrainingLoss(VAELossType,
                           [0.1,            # KLD
                            5,              # Angle
                            0.02,           # FK
                            50.,            # PosVel
                            1.,             # Rotflat
                            5.,             # RotVelVert
                            2.,             # Height
                            0.1,            # Feet
                            2.,             # Latent Smooth
                            2,              # Angle Smooth
                            0.01,           # FK Smooth
                            1.,             # PosVel Smooth
                            2.,             # Rotflat Smooth
                            0.,             # RotVelVert Smooth
                            1.,             # Height Smooth
                            0.05,           # FeetSmooth
                            0.5 ]           # Contacts
                            )

training_config:TrainingConfig = TrainingConfig(20, 50, 0.001, 0.8)

Train(training_config, error_config, "VAE/VAE_"+name+".pymodel", "VAE_"+name+time.strftime("_%m-%d-%H:%M"), DB_train, DB_validate, visual = True)
```
